```python


# 2. Import
from ultralytics import YOLO
import cv2

# 3. Load model (auto-downloads)
model = YOLO("yolov8n.pt")

# 4. Load image
img = cv2.imread("image.jpg")

# 5. Run detection
results = model(img)

# 6. Show result
output = results[0].plot()
cv2.imshow("Detection", output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    results = model(frame)
    output = results[0].plot()

    cv2.imshow("Detection", output)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
